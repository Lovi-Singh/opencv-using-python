```python
## Plotting Histogram Of Image

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

ir_image = cv2.imread('../Desktop/model/Thermal images/FLIR1239_crpd.jpg')
show_img = cv2.cvtColor(ir_image, cv2.COLOR_BGR2RGB)

plt.imshow(show_img)

img = ir_image

color = ('b','g','r')
plt.figure(figsize=(20,15))
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.title('Histogram for Infrared Image',fontsize=20)
    plt.xlim([0,256])
    plt.xlabel("Intensity values",fontsize=20)
    plt.ylabel("Number of pixels", fontsize=20)
    plt.plot(histr,color=col)
    

```
