```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img1 = cv2.imread("../Desktop/Computer-Vision-with-Python/DATA/00-puppy.jpg", 0)

img2 = cv2.imread("../Desktop/Computer-Vision-with-Python/DATA/puppy -edited.jpg" , 0)
```

```python
plt.imshow(img1,cmap='gray')
```

```python
plt.imshow(img2,cmap='gray')
```

```python
sub = cv2.subtract(img1,img2)
sub
```

```python
ret , threshub = cv2.threshold(sub,0.5,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(threshub, cmap = 'gray')
```

```python
frame_diff = cv2.absdiff(img2,img1)
img_rsz = cv2.resize(frame_diff,(1300,800))
while True:
    cv2.imshow('frame difference resized', img_rsz)

    if cv2.waitKey(1) & 0xFF == 27:
         break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img1 = cv2.imread("../Pictures/nbwopen.jpg", 0)

img2 = cv2.imread("../Pictures/nbwpen.jpg" , 0)

frame_diff = cv2.absdiff(img2,img1)

plt.imshow(frame_diff,cmap='gray')
```

```python
ret , threshub = cv2.threshold(frame_diff,110,255,cv2.THRESH_BINARY)
ret
```

```python
plt.imshow(threshub,cmap='gray')
```

```python
kernel = np.ones((3,3),dtype=np.uint8)
```

```python
result = cv2.erode(threshub,kernel,iterations=1)
```

```python
plt.imshow(result,cmap='gray')
```
